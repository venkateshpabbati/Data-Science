```python
#load the libraries
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
raw_data = pd.read_csv("crime_data.csv")
raw_data.head()
```

```python
raw_data.rename(columns={'Unnamed: 0':'city'}, inplace=True)
raw_data.head()
```

```python
raw_data.info()
```

```python
raw_data.describe()
```

```python
sns.boxplot(data=raw_data)
```

```python
# Extracting features for clustering
array = raw_data.iloc[:,1:]

# Standardizing the data using StandardScaler
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_data = scaler.fit_transform(array)
scaled_data
```

```python
# Normalizing the data using MinMaxScaler
from sklearn.preprocessing import MinMaxScaler
norm_scaler = MinMaxScaler(feature_range=(0,1))
normscaled = norm_scaler.fit_transform(array)
```

```python
normscaled
```

```python
from sklearn.cluster import KMeans

# Finding the optimal number of clusters using the Elbow Method
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, random_state=0)
    kmeans.fit(scaled_data)
    wcss.append(kmeans.inertia_)
    
plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
# Number of clusters (optimal) = 4
clusters_kmeans = KMeans(n_clusters=4, random_state=5)
clusters_kmeans.fit(scaled_data)
```

```python
clusters_kmeans.labels_
```

```python
#Assigning clusters to the datset
data_kmeans = raw_data.copy()
data_kmeans['clusters_kmeans'] = clusters_kmeans.labels_
```

```python
clusters_kmeans.cluster_centers_
```

```python
data_kmeans.groupby('clusters_kmeans').mean(numeric_only=True)
```

```python
data_kmeans.head()
```

```python
#Visualising the data via dimensionality reduction
from sklearn.decomposition import PCA

pca = PCA(n_components=2).fit(scaled_data)
pca_2d = pca.transform(scaled_data)
```

```python
plt.scatter(
    x=pca_2d[:, 0],  # Select first column for x-axis
    y=pca_2d[:, 1],  # Select second column for y-axis
    c=data_kmeans.clusters_kmeans,  # Color points based on cluster assignment
    marker="o",  # Use the same marker for all points
    cmap="Accent",  # Use a color map for clusters (adjust as needed)
)

plt.title('Crime dataset with K-Means Clustering')
plt.colorbar(label="Cluster")  # Add colorbar with appropriate label
plt.show()
```

```python
#Agglomerative Hierarchical Clustering
```

```python
import scipy.cluster.hierarchy as sch

#dendrogram = sch.dendrogram(sch.linkage(normscaled, method='single'))

# Assuming distance matrix is available in 'distance_matrix' variable
dendrogram = sch.linkage(normscaled, method='single')
plt.figure(figsize=(10, 6))  # Adjust figure size as needed
plt.title('Dendrogram of Hierarchical Clustering')
sch.dendrogram(dendrogram)
plt.show()
```

```python
# Assuming distance matrix is available in 'distance_matrix' variable
dendrogram = sch.linkage(normscaled, method='ward')
plt.figure(figsize=(10, 6))  # Adjust figure size as needed
plt.title('Dendrogram of Hierarchical Clustering')
sch.dendrogram(dendrogram)
plt.show()
```

```python
# Assuming distance matrix is available in 'distance_matrix' variable
dendrogram = sch.linkage(normscaled, method='complete')
plt.figure(figsize=(10, 6))  # Adjust figure size as needed
plt.title('Dendrogram of Hierarchical Clustering')
sch.dendrogram(dendrogram)
plt.show()
```

```python
# Assuming distance matrix is available in 'distance_matrix' variable
dendrogram = sch.linkage(normscaled, method='average')
plt.figure(figsize=(10, 6))  # Adjust figure size as needed
plt.title('Dendrogram of Hierarchical Clustering')
sch.dendrogram(dendrogram)
plt.show()
```

```python
# Assuming distance matrix is available in 'distance_matrix' variable
dendrogram = sch.linkage(normscaled, method='centroid')
plt.figure(figsize=(10, 6))  # Adjust figure size as needed
plt.title('Dendrogram of Hierarchical Clustering')
sch.dendrogram(dendrogram)
plt.show()
```

```python
# Assuming distance matrix is available in 'distance_matrix' variable
dendrogram = sch.linkage(normscaled, method='median')
plt.figure(figsize=(10, 6))  # Adjust figure size as needed
plt.title('Dendrogram of Hierarchical Clustering')
sch.dendrogram(dendrogram)
plt.show()
```

```python
# Assuming distance matrix is available in 'distance_matrix' variable
dendrogram = sch.linkage(normscaled, method='weighted')
plt.figure(figsize=(10, 6))  # Adjust figure size as needed
plt.title('Dendrogram of Hierarchical Clustering')
sch.dendrogram(dendrogram)
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
# Performing Agglomerative Clustering with 4 clusters
hc = AgglomerativeClustering(n_clusters=4, linkage='complete')
y_hc = hc.fit_predict(normscaled)
```

```python
y_hc
```

```python
data_agglo = raw_data.copy()
data_agglo['clusters_agglo'] = y_hc
data_agglo['clusters_agglo'].value_counts()
```

```python
#model evaluation
data_agglo.groupby('clusters_agglo').mean(numeric_only=True)
```

```python
#Visualising the data via dimensionality reduction

pca_agg = PCA(n_components=2).fit(normscaled)
pca_2d_agg = pca_agg.transform(normscaled)
```

```python
import matplotlib.pyplot as plt

# Assuming 'pca_2d_agg' contains the transformed data and 'data_agglo.clusters_agglo' holds cluster labels

# Define colors and markers for each cluster (adjust as needed)
colors = ['r', 'g', 'b', 'm']
markers = ['+', 'o', '*', 'v']

# Create the scatter plot with separate coloring and markers for clusters
plt.figure(figsize=(8, 6))  # Adjust figure size as desired
for i in range(len(data_agglo.clusters_agglo)):
    cluster_label = data_agglo.clusters_agglo[i]
    plt.scatter(pca_2d_agg[i, 0], pca_2d_agg[i, 1], c=colors[cluster_label], marker=markers[cluster_label])

# Add labels and title
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('Crime dataset with Hierarchical Agglomerative Clustering')
plt.tight_layout()  # Adjust layout to prevent overlapping elements
plt.show()
```

```python
from sklearn.cluster import DBSCAN

dbscan = DBSCAN(eps=1, min_samples=4)
dbscan.fit(scaled_data)
```

```python
dbscan.labels_
```

```python
# model evaluation
data_dbscan.groupby('clusters_dbscan').mean(numeric_only=True)
```
