```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("book.csv", encoding="ISO-8859-1").rename(columns={'User.ID': 'userid', 'Book.Title': 'title', 'Book.Rating': 'rating'})
```

```python
df[0:5]
```

```python
#number of unique users in the dataset
len(df.userid.unique())
```

```python
len(df.title.unique())
```

```python
books_df = df.pivot_table(index='userid',columns='title',values='rating', fill_value=0).reset_index(drop=True)
books_df
```

```python
books_df
```

```python
books_df.index = df.userid.unique()
```

```python
books_df
```

```python
#Impute those NaNs with 0 values
books_df.fillna(0, inplace=True)
```

```python
books_df
```

```python
#Calculating Cosine Similarity between Users
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```python
user_sim = 1 - pairwise_distances( books_df.values,metric='cosine')
```

```python
user_sim
```

```python
#Store the results in a dataframe
user_sim_df = pd.DataFrame(user_sim)
```

```python
#Set the index and column names to user ids 
user_sim_df.index = df.userid.unique()
user_sim_df.columns = df.userid.unique()
```

```python
user_sim_df.iloc[0:5, 0:5]
```

```python
np.fill_diagonal(user_sim, 0)
user_sim_df.iloc[0:5, 0:5]
```

```python
#Most Similar Users
user_sim_df.idxmax(axis=1)[0:5]
```
