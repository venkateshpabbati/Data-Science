```python
def expected_value(values, weights):
    values = np.asarray(values)
    weights = np.asarray(weights)
    return (values*weights).sum() / weights.sum()
```

```python
c_count = [1,4,3,5,6,2]
ch_prob = [0.015,0.20,0.65,0.005,0.01,0.120]
expected_value(c_count, ch_prob)
```

```python
cars = pd.read_csv('Cars.csv')
```

```python
Prob_MPG_greater_than_38 = 1 - stats.norm.cdf(38, loc= cars.MPG.mean(), scale= cars.MPG.std())
print('P(MPG>38)=',Prob_MPG_greater_than_38)
```

```python
prob_MPG_less_than_40 = stats.norm.cdf(40, loc=cars.MPG.mean(),scale=cars.MPG.std())
print('P(MPG<40)=', prob_MPG_less_than_40)
```

```python
Prob_MPG_less_than_20 = stats.norm.cdf(20, loc= cars.MPG.mean(), scale= cars.MPG.std())
print('P(MPG<20)=',Prob_MPG_less_than_20)
```

```python
Prob_MPG_less_than_50 = stats.norm.cdf(50, loc= cars.MPG.mean(), scale= cars.MPG.std())
print('P(MPG<50)=',Prob_MPG_less_than_50)
```

```python
prob_20_lessthan_MPG_lessthan_50  = Prob_MPG_less_than_50-Prob_MPG_less_than_20
print('P(20<MPG<50)=', prob_20_lessthan_MPG_lessthan_50)
```
