# Feature Engineering

```python
#Encoding

from sklearn.preprocessing import LabelEncoder

labels = ['workclass', 'maritalstatus', 'occupation', 'relationship', 'race', 'sex','native', 'Salary']
train_en = train.copy()
test_en = test.copy()
le = LabelEncoder()
for i in labels:
    train_en[i] = le.fit_transform(train_en[i])
    test_en[i] = le.fit_transform(test_en[i])
```

```python
train_en.head()
```

```python
#Split the data 

X_train = train_en.iloc[:, :-1]
y_train = train_en['Salary']

X_test = test_en.iloc[:, :-1]
y_test = test_en['Salary']
```

# build the model

```python
from sklearn.naive_bayes import GaussianNB, MultinomialNB
```

```python
#GuassianNB
gnb = GaussianNB()
model_gnb = gnb.fit(X_train, y_train)
model_gnb
```

```python
#MultinommialNB
mnb = MultinomialNB()
model_mnb = mnb.fit(X_train, y_train)
model_mnb
```

# Evaluate the model

```python
from sklearn.metrics import classification_report
```

```python
y_pred_gnb = model_gnb.predict(X_test)
print(classification_report(y_test, y_pred_gnb))
```

```python
y_pred_mnb = mnb.predict(X_test)
print(classification_report(y_test, y_pred_mnb))
```

```python
# use cross validation techniques
```

```python
from sklearn.model_selection import KFold
```

```python
from sklearn.model_selection import KFold, cross_val_score

# Initialize KFold
kfold = KFold(n_splits=5, shuffle=True, random_state=42)  # Define the number of splits and random state

# Perform k-fold cross-validation for Multinomial Naive Bayes
multinomial_cv_results = cross_val_score(model_mnb, X_train, y_train, cv=kfold)

# Perform k-fold cross-validation for Gaussian Naive Bayes
gaussian_cv_results = cross_val_score(model_gnb, X_train, y_train, cv=kfold)

# Display the cross-validation results
print("Multinomial Naive Bayes Cross-validation scores:", multinomial_cv_results)
print("Multinomial Naive Bayes Mean cross-validation score:", multinomial_cv_results.mean())

print("\nGaussian Naive Bayes Cross-validation scores:", gaussian_cv_results)
print("Gaussian Naive Bayes Mean cross-validation score:", gaussian_cv_results.mean())
```
